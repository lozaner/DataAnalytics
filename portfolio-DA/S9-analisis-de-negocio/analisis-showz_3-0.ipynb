 ### 1. ¿Cuánto dinero se gastó?  (Total/por fuente de adquisición/a lo largo del tiempo) 

```python
# Asegurar que la columna de fecha es de tipo datetime
cost_raw['dt'] = pd.to_datetime(cost_raw['dt'])

#  Calcular el gasto total
total_cost = cost_raw['costs'].sum()

#  Calcular el gasto por fuente de adquisición
cost_by_source = cost_raw.groupby('source_id')['costs'].sum().sort_values(ascending=False).reset_index()

#  Calcular el gasto a lo largo del tiempo (por día, mes o año)
cost_by_date = cost_raw.groupby(cost_raw['dt'].dt.to_period('M'))['costs'].sum().reset_index()

# Mostrar los resultados
print(f"Total gastado: {total_cost}")
print('----------------------------------')
print("\nGasto por fuente de adquisición:")
print(cost_by_source)
print('----------------------------------')
print("\nGasto a lo largo del tiempo:")
print(cost_by_date)

```

### 3. ¿Cuán rentables eran las inversiones? (ROMI)

```python
# INSTRUCCIONES BASE:
# Debes tener para cada usuario su source de primera visita, mes de primera compra y todas sus compras. 
# Es decir, una tabla con una fila para cada compra y en cada fila tener el source de la 1era visita 
# y mes de 1era compra del usuario correspondiente.

#   - Este primer dataframe lo puedes construir obteniendo la primera visita de cada usuario, eso te da la source de la primera visita
#   - Luego obtén la fecha de primera compra para cada usuario
# ESTOS DOS PASOS YA SE REALIZARON EN LA PARTE DEL CAC: primera_data_usuarios CONTIENE ESA DATA
#   - Haz un par de left joins para unir esta data a orders, así tendrás lo requerido
```

```python
# --- PASO 0: Asegurar las columnas de fecha y mes necesarias ---

# En orders_raw: Asegurar datetime y crear mes del pedido
orders_raw['buy_ts'] = pd.to_datetime(orders_raw['buy_ts'])
orders_raw['order_month'] = orders_raw['buy_ts'].dt.to_period('M')

# En cost_raw: Asegurar datetime y crear mes del costo
cost_raw['dt'] = pd.to_datetime(cost_raw['dt'])
cost_raw['costs_month'] = cost_raw['dt'].dt.to_period('M')

# En primeras_compras: Asegurar datetime y crear mes de primera compra
# (Asegúrate de tener 'primeras_compras' como en tu bloque [66])
primeras_compras['primera_compra_ts'] = pd.to_datetime(primeras_compras['primera_compra_ts'])
primeras_compras['primera_compra_mes'] = primeras_compras['primera_compra_ts'].dt.to_period('M')

# DataFrame base: Todas las órdenes con la fuente de adquisición original y el mes de la primera compra
# Necesitamos: primera_fuente_usuarios (uid, source_id) y primeras_compras (uid, primera_compra_mes)
# (Asegúrate de que 'primera_fuente_usuarios' exista como en tu bloque [93])
orders_with_acquisition_info = pd.merge(orders_raw, primera_fuente_usuarios, on='uid', how='left')
orders_with_acquisition_info = pd.merge(orders_with_acquisition_info,
                                       primeras_compras[['uid', 'primera_compra_mes']], # Solo necesitamos estas columnas
                                       on='uid',
                                       how='left')

# Agrupar costos mensuales por fuente para unir más tarde
monthly_costs_agg = cost_raw.groupby(['source_id', 'costs_month']).agg(monthly_total_costs=('costs', 'sum')).reset_index()

# Obtener lista única de fuentes para iterar
sources_to_process = orders_with_acquisition_info['source_id'].dropna().unique()
```

```python
print("Iniciando cálculo de ROMI por fuente...")

# --- PASO 1: Iterar sobre cada fuente ---
for source_i in sources_to_process:
    source_i = int(source_i) # Asegurar que sea entero si es necesario
    print(f"\nProcesando Fuente: {source_i}")

    # Omitir fuente 7 como en tu código original
    if source_i == 7:
        print(f"Omitiendo Fuente {source_i} por datos insuficientes.")
        continue

    # Filtrar datos de órdenes para la fuente actual
    source_data = orders_with_acquisition_info[orders_with_acquisition_info['source_id'] == source_i].copy()

    if source_data.empty:
        print(f"No hay datos de órdenes para la Fuente {source_i}.")
        continue

    # --- PASO 2: Calcular Tamaño de Cohortes (para esta fuente) ---
    cohort_sizes = source_data.groupby('primera_compra_mes').agg(n_buyers=('uid', 'nunique')).reset_index()

    if cohort_sizes.empty:
        print(f"No se pudieron calcular tamaños de cohortes para la Fuente {source_i}.")
        continue

    # --- PASO 3: Calcular Ingresos por Cohorte y Mes (para esta fuente) ---
    cohort_revenue = source_data.groupby(['primera_compra_mes', 'order_month']).agg(monthly_revenue=('revenue', 'sum')).reset_index()

    # --- PASO 4: Combinar y Calcular LTV Mensual ---
    report_ltv = pd.merge(cohort_revenue, cohort_sizes, on='primera_compra_mes', how='left')
    report_ltv['age'] = ((report_ltv['order_month'] - report_ltv['primera_compra_mes']).apply(lambda x: x.n)).astype(int) # Edad en meses enteros
    report_ltv['ltv'] = report_ltv['monthly_revenue'] / report_ltv['n_buyers']

    # --- PASO 5: Incorporar Costos y Calcular CAC ---
    # Unir con costos agregados, haciendo coincidir el mes de la primera compra con el mes del costo
    report_full = pd.merge(report_ltv,
                           monthly_costs_agg[monthly_costs_agg['source_id'] == source_i],
                           left_on='primera_compra_mes',
                           right_on='costs_month',
                           how='left') # Usar left join por si falta algún costo

    # Eliminar columnas redundantes si es necesario (source_id_y, costs_month)
    if 'source_id_y' in report_full.columns:
       report_full = report_full.drop(columns=['source_id_y'])
    if 'costs_month' in report_full.columns:
       report_full = report_full.drop(columns=['costs_month'])
    # Renombrar source_id_x si existe
    if 'source_id_x' in report_full.columns:
       report_full = report_full.rename(columns={'source_id_x': 'source_id'})


    # Calcular CAC (Costo del mes de adquisición / Compradores de esa cohorte)
    # Llenar NaN en costos con 0 podría ser necesario si algunos meses no tuvieron gasto
    report_full['monthly_total_costs'] = report_full['monthly_total_costs'].fillna(0)
    # Evitar división por cero si n_buyers es 0 (aunque no debería pasar si se calcula bien)
    report_full['cac'] = np.where(report_full['n_buyers'] > 0,
                                 report_full['monthly_total_costs'] / report_full['n_buyers'],
                                 0)

    # --- PASO 6: Calcular ROMI Mensual ---
    # Evitar división por cero si CAC es 0
    report_full['romi'] = np.where(report_full['cac'] > 0,
                                   report_full['ltv'] / report_full['cac'],
                                   0) # O podrías poner np.nan o un valor grande si LTV > 0 y CAC = 0

    # --- PASO 7: Pivotar y Calcular ROMI Acumulado ---
    try:
        romi_pivot = report_full.pivot_table(index='primera_compra_mes',
                                             columns='age',
                                             values='romi',
                                             aggfunc='mean') # mean por si acaso, aunque no debería haber duplicados

    # Calcular ROMI acumulado
        romi_acumulative = romi_pivot.cumsum(axis=1)

    except Exception as e:
        print(f"Error al crear tabla pivot o calcular cumsum para Fuente {source_i}: {e}")
        continue # Saltar a la siguiente fuente si hay error

    
    # --- PASO 8: Visualizar ---
    plt.figure(figsize=(13, 9))
    sns.heatmap(romi_acumulative,
                annot=True,
                fmt='.2f',
                linewidths=0.5, # Ajustado de 1 a 0.5 para mejor visualización a veces
                linecolor='grey',
                # cbar_kws={'orientation': 'horizontal'} # Puedes descomentar si prefieres barra horizontal
               ).set(title=f'ROMI Acumulado por Cohorte - Fuente {source_i}')
    plt.xlabel("Edad (Meses)")
    plt.ylabel("Mes de Primera Compra (Cohorte)")
    plt.show()

print("\nCálculo de ROMI por fuente finalizado.")
```
